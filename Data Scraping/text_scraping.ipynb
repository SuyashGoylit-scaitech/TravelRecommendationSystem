# Destination scraping

```python
# Testing the authenticity of the urls (Change this code to markdown, once executed, to save time)

for i in destinations:
    url = "https://www.holidify.com/places/"+i
    P_link = requests.get(url)
    print(link)

    html = link.text
    soup = BeautifulSoup(html, "html.parser")
```

```python
column_set1 = ['Place', 'State', 'Rating', 'Ideal Time to Visit', 'Ideal duration', 'Nearest Airport', 'Starting Budget']
column_set2 = ['Main Attraction', 'About', 'Top Places to Visit']
data_head = pd.DataFrame(columns = column_set1)
data_body = pd.DataFrame(columns = column_set2)
```

```python
for i in destinations:
    url = "https://www.holidify.com/places/"+i
    link = requests.get(url)
    #print(link)

    html = link.text
    soup = BeautifulSoup(html, "html.parser")
    
    containers_head = soup.findAll("div", {"class" : "col-12 col-lg-6 right"})
    containers_body = soup.findAll("div", {"class" : "col-lg-8 pr-lg-2"})
    
    container_head = containers_head[0]
    container_body = containers_body[0]

    # For container_head in containers_head
    
    name = container_head.findAll('h1')[0].get_text().strip()
    
    state = container_head.findAll('div', {'class': 'mb-2 font-smaller'})[0].get_text()
    state = state.replace('\n', '').strip()
    
    rating = container_head.findAll('span', {'class': 'rating-badge'})[0].get_text().strip()
    rating = rating[:3]
    
    time = container_head.findAll('div', {'class': 'col-12 col-md-6'})[1].get_text()
    time = time.replace('\n', '')
    time = time[11:]
    time = time[:-10]
    time = time.strip()
    
    duration = container_head.findAll('div', {'class': 'col-12 col-md-6'})[2].get_text()
    duration = duration.replace('\n', '')
    duration = duration[16:].strip()
    
    if container_head.findAll('div', {'class': 'col-12 col-md-6'})[3]:
        airport = container_head.findAll('div', {'class': 'col-12 col-md-6'})[3].get_text()
        airport = airport.replace('\n', '')
        airport = airport[18:]
        airport = airport[:-15]
        airport = airport.strip()
    
    if container_head.findAll('span', {'class': 'price-large'}):
        budget = container_head.findAll('span', {'class': 'price-large'})[0].get_text().strip()
    
    Data = pd.DataFrame([[name , state , rating, time , duration, airport, budget]])
    Data.columns = column_set1
    data_head = data_head.append(Data, ignore_index = True)
        
    
    # For container_body in containers_body
    
    if container_body.findAll('h3', {'class': 'tagline w-100'}):
        tagline = container_body.findAll('h3', {'class': 'tagline w-100'})[0].get_text().strip()
    else:
        tagline = ''
    
    about = container_body.findAll('div', {'class': 'readMoreText'})[0].get_text().strip()
    
    num = len(container_body.findAll('div', {'class': 'col-6 col-md-4 ptv-item'}))
    attraction = ''
    for j in range(0, num):
        attraction = attraction +  ', ' + container_body.findAll('div', {'class': 'col-6 col-md-4 ptv-item'})[j].get_text()
    attraction = attraction.replace('\n', '')
    attraction = attraction[2:].strip()
    
    Data = pd.DataFrame([[tagline, about, attraction]])
    Data.columns = column_set2
    data_body = data_body.append(Data, ignore_index = True)
```

```python
data = pd.merge(data_head, data_body, left_index=True, right_index=True)
data.to_excel('Destinations_checkpoint.xlsx', index = False)
```

```python
# Fixing the missing budget values

data['Starting Budget'] = data['Starting Budget'].astype('string')
data['temp'] = data['Starting Budget'].apply(lambda x: x[0])

data['Starting Budget'] = data['Starting Budget'].apply(lambda x: x[1:])
data.loc[data['temp']!='₹', ['Starting Budget']] = '0'
data['Starting Budget'] = data['Starting Budget'].apply(lambda x: x.replace(',', ''))
data['Starting Budget'] = data['Starting Budget'].astype('int')

# data[data['Starting Budget'] == 0].value_count() = 74

data['Starting Budget'] = data['Starting Budget'].replace(0, data['Starting Budget'].sum() / 74)

data
```

```python
data['Starting Budget'] = data['Starting Budget'].astype('str')
data['Starting Budget'] = '₹' + data['Starting Budget']
data.drop(['temp'], axis = 1, inplace = True)
data.to_excel('Destinations.xlsx', index = False)
```

```python
# tsv conversion for backend

data.drop(['Main Attraction'], axis = 1, inplace = True)
data.to_csv('Destinations.csv', sep = '\t', index = False)
```

# Hotel scraping

```python
# Testing the authenticity of the urls (Change this code to markdown, once executed, to save time)

for i in destinations:
    url = "https://www.holidify.com/places/" + i + "/hotels-where-to-stay.html"
    link = requests.get(url)
    print(link)

    html = link.text
    soup = BeautifulSoup(html, "html.parser")
```

```python
column_set = ['Destination', 'Hotel', 'Starting Budget', 'Distance from city centre', 'Rating']
hotels = pd.DataFrame(columns = column_set)
```

```python
for i, j in zip(destinations, destinations_raw):
    url = "https://www.holidify.com/places/" + i + "/hotels-where-to-stay.html"
    link = requests.get(url)
    html = link.text
    soup = BeautifulSoup(html, "html.parser")
    
    containers = soup.findAll('div', class_= 'row no-gutters mb-3 hotel-card')
    
    if containers:
        count = 6
        
        if len(containers) < 6:
            count = len(containers)
    
        k = 0
    
        while k < count:
            container = containers[k]
            
            destination = j
            
            hotel = container.findAll('h5', class_= 'hotel-name mt-md-3 hotel-line-margin noMargin')[0].get_text().strip()
            
            price = container.findAll('p', class_= 'price')[0].get_text().strip()
            
            distance = container.findAll('p', class_='hotel-line-margin')[0].get_text().strip()
            
            if container.findAll('span', class_= 'rating-badge'):
                rating = container.findAll('span', class_= 'rating-badge')[0].get_text().strip()
            
            Data = pd.DataFrame([[destination, hotel, price, distance, rating]])
            Data.columns = column_set
            hotels = hotels.append(Data, ignore_index = True)
            
            k += 1
```
